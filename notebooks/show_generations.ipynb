```python
import pandas as pd
import os

CSV_PATH = '../results/generated_text/all_generations.csv'
BEST_MODEL = 'RNN'
BEST_OPTIMIZER = 'RMSprop'

def print_fancy_header(text):
    print("\n" + "="*80)
    print(f"  {text}")
    print("="*80 + "\n")

def main():
    # 1. Load Data
    if not os.path.exists(CSV_PATH):
        print(f"โ Error: File not found at {CSV_PATH}")
        return

    df = pd.read_csv(CSV_PATH)

    
    winner_df = df[(df['Model'] == BEST_MODEL) & (df['Optimizer'] == BEST_OPTIMIZER)]

    if winner_df.empty:
        print("โ๏ธ No data found for the best model combination.")
        return

    print_fancy_header(f"๐ BEST MODEL GENERATIONS: {BEST_MODEL} + {BEST_OPTIMIZER}")
    
    print(f"{'SEED':<10} | {'TEMP':<5} | {'GENERATED POETRY (URDU)':<50}")
    print("-" * 80)

    for index, row in winner_df.iterrows():
        seed = row['Seed']
        temp = row['Temperature']
        text = row['Generated Text']
        
        text = text.replace('\n', ' ')
        
        if len(text) > 80:
            text = text[:77] + "..."

        print(f"{seed:<10} | {temp:<5} | {text}")
    
    print("-" * 80)
    print("\n")

    print_fancy_header("๐ CROSS-MODEL COMPARISON (Seed: 'ูุญุจุช', Temp: 1.0)")
    
    comparison_df = df[(df['Seed'] == 'ูุญุจุช') & (df['Temperature'] == 1.0)]
    
    print(f"{'MODEL':<12} | {'OPTIMIZER':<10} | {'GENERATED TEXT'}")
    print("-" * 80)
    
    for index, row in comparison_df.iterrows():
        model = row['Model']
        opt = row['Optimizer']
        text = row['Generated Text']
        if len(text) > 60: text = text[:57] + "..."
        
        print(f"{model:<12} | {opt:<10} | {text}")
        
    print("-" * 80)

if __name__ == "__main__":
    main()
```
